## Exercício 3.2 — Observações (teste falhando propositalmente)

- **1. O pipeline ficou verde ou vermelho?**  
  Vermelho quando existe ao menos 1 teste falhando.

- **2. Qual foi a mensagem de erro na seção FAILURES?**  
  Um `AssertionError` do tipo `assert 2 == 3` (ou equivalente), apontando a linha do teste.

- **3. O pytest parou no primeiro teste que falhou ou continuou rodando?**  
  Por padrão, continua rodando e reporta o resumo; para parar no primeiro, usa-se `-x`.

- **4. Se houver jobs com `needs` e o job de testes falhar, o que acontece com o job dependente?**  
  O job dependente não executa e aparece como *skipped*.
